## Lorenz system

```python
# Define the Lorenz system
def lorenz(t, state, sigma, rho, beta):
    x, y, z = state
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

# Parameters for the Lorenz system
sigma, rho, beta = 10.0, 28.0, 8.0 / 3.0  # Typical chaotic values
initial_state_lorenz = [1.0, 1.0, 1.0]  # Initial condition
t_span_lorenz = (0, 50)  # Time range for simulation
t_eval_lorenz = np.linspace(t_span_lorenz[0], t_span_lorenz[1], 10000)  # Time points for output

# Solve the system
solution_lorenz = solve_ivp(lorenz, t_span_lorenz, initial_state_lorenz, 
                            args=(sigma, rho, beta), t_eval=t_eval_lorenz, method='RK45')

# Extract the solution
x_lorenz, y_lorenz, z_lorenz = solution_lorenz.y

# Plot the phase space and time series
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Phase space plot (x vs. z)
axs[0].plot(x_lorenz, z_lorenz, color='blue', lw=0.7)
axs[0].set_title('Lorenz Phase Space (x vs. z)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('z')
axs[0].grid()

# Time series plot for x
axs[1].plot(solution_lorenz.t, x_lorenz, color='red', lw=0.7)
axs[1].set_title('Lorenz Time Series (x vs. t)')
axs[1].set_xlabel('Time (t)')
axs[1].set_ylabel('x')
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
# Create a 3D plot for the Lorenz system

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the Lorenz attractor in 3D (x, y, z)
ax.plot(x_lorenz, y_lorenz, z_lorenz, lw=0.7, color='blue')
ax.set_title('3D Plot of the Lorenz Attractor')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

## Rossler system

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the Rossler system
def rossler(t, state, a, b, c):
    x, y, z = state
    dxdt = -y - z
    dydt = x + a * y
    dzdt = b + z * (x - c)
    return [dxdt, dydt, dzdt]

# Parameters for the Rossler system
a, b, c = 0.2, 0.2, 5.7  # Typical values for chaotic behavior
initial_state = [1.0, 1.0, 1.0]  # Initial condition
t_span = (0, 100)  # Time range for simulation
t_eval = np.linspace(t_span[0], t_span[1], 10000)  # Time points for output

# Solve the system
solution = solve_ivp(rossler, t_span, initial_state, args=(a, b, c), t_eval=t_eval, method='RK45')

# Extract the solution
x, y, z = solution.y

# Plot the phase space and time series
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Phase space plot (x vs. z)
axs[0].plot(x, z, color='blue', lw=0.7)
axs[0].set_title('Phase Space (x vs. z)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('z')
axs[0].grid()

# Time series plot for x
axs[1].plot(solution.t, x, color='red', lw=0.7)
axs[1].set_title('Time Series (x vs. t)')
axs[1].set_xlabel('Time (t)')
axs[1].set_ylabel('x')
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
# Create a 3D plot for the Rossler system

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the Rossler attractor in 3D (x, y, z)
ax.plot(x, y, z, lw=0.7, color='blue')
ax.set_title('3D Plot of the Rssler Attractor')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
